# 思维导图

```python
#!pip install pyvis

from pyvis.network import Network

def create_streamlined_mind_map():
    nt = Network(notebook=True)
    nt.add_node("SEO")

    platforms = ["Yahoo Platform for SEO", "Bing Platform for SEO"]
    yahoo_subtopics = ["Yahoo Keyword Research", "Yahoo Backlinks", "Yahoo Analytics Integration"]
    bing_subtopics = ["Bing Webmaster Tools", "Bing Ads", "Bing Organic Traffic Analysis"]

    for platform in platforms:
        nt.add_node(platform, title=platform)
        nt.add_edge("SEO", platform)

    for sub in yahoo_subtopics:
        nt.add_node(sub, title=sub)
        nt.add_edge("Yahoo Platform for SEO", sub)

    for sub in bing_subtopics:
        nt.add_node(sub, title=sub)
        nt.add_edge("Bing Platform for SEO", sub)

    return nt

nt = create_streamlined_mind_map()

```

# 表格图表

```python
import matplotlib.pyplot as plt

# 设置图形大小
fig, ax = plt.subplots(figsize=(12, 8))

# 设置标题
ax.set_title('Projects Overview')

# 根据项目类别和其数量来定位和显示每个项目
y = 0
for category, project_list in projects.items():
    for project in project_list:
        ax.text(0, y, project["title"], weight='bold', size=10)
        ax.text(1, y, project["date"], size=10)
        ax.text(2, y, ', '.join(project["key_points"]), size=10, wrap=True)
        y -= 1  # 为下一个项目移动Y坐标位置

    y -= 1  # 在不同的项目类别之间添加一些空间

# 设置坐标轴的标签和标题
ax.set_yticks([])  # 隐藏y坐标
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["Title", "Date", "Key Points"])
ax.axis('off')  # 隐藏坐标轴

plt.tight_layout()
plt.show()
```
